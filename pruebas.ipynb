## trabajo con variables categoricas los valores unicos que contiene

- crear un set de datos que contega solo los valores unicos

```python
paises = data['País']
paises = np.array(paises)

def funcion_paises_unicos(lista_paises):
    
    lista_paises_unicos = []
    
    paises_unicos = set(lista_paises)
    
    for pais in paises_unicos:
        lista_paises_unicos.append(pais)
        
    return(lista_paises_unicos)
        

unicos_paises = funcion_paises_unicos(paises)
print(unicos_paises)
```
